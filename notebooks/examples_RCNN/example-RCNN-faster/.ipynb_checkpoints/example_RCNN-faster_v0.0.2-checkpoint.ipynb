```python
#Disable progress bar to avoid requiring widget
tfds.disable_progress_bar()

#Instantiate the builder
dsID = 'coco/2017'
builder = tfds.builder(dsID)
```

```python
builder.info.splits
```

```python
print('data_dir', builder.data_dir)
builder.info
```

```python
warnings.filterwarnings('ignore')
builder.download_and_prepare()
warnings.filterwarnings('default')
```

```python
# fetch text label IDs
labels = list(pd.read_csv(os.path.join(builder.data_dir,'objects-label.labels.txt'), header=None).iloc[:,0])
print('n labels:', len(labels))
labels
```

# Prepare the Data

```python
#Define some constants
batch_size = 32
n_classes = len(labels)
print('n_classes',n_classes)
```

```python
#Prepare the data generator
# ds = builder.as_dataset(batch_size=batch_size)
# ds_train_gen = tfds.as_numpy(ds['train'])
# ds_test_gen = tfds.as_numpy(ds['test'])

#slice out a subset of the data for development
slice_ = '[:10%]'

#Don't apply a batch size with tfds. It will make all images in a batch the same size and not adjust the bounding box coordinates accordingly
ds_train = tfds.load(dsID, split='train'+slice_)

ds_test = tfds.load(dsID, split='validation'+slice_)

n_training_examples = len(list(ds_train))
n_testing_examples = len(list(ds_test))

print('n_training_examples:', n_training_examples)
print('n_testing_examples:', n_testing_examples)
```

# Map the Augmentation Function

```python
augment_args = {'dtype': tf.float32,
                'flip_left_right':True, 
                'flip_up_down':True, 
                'rot90':True, 
                'smallest_side_length':300, 
                'verbose':0 }

ds_train = ds_train.map(lambda ds_slice: NN.RCNN.augment(split = 'train', ds_slice = ds_slice,
                                                          **augment_args))
ds_train_gen = tfds.as_numpy(ds_train)

ds_test = ds_test.map(lambda ds_slice: NN.RCNN.augment(split='test', ds_slice = ds_slice,
                                                        **augment_args))
ds_test_gen = tfds.as_numpy(ds_test)
```

# Visualize The Data

```python
for i in range(3):
    ds_slice = next(ds_train_gen)
    
    NN.RCNN.plot.img_and_bboxes(ds_slice, labels=labels)
    
    ds_slice = next(ds_test_gen)
    NN.RCNN.plot.img_and_bboxes(ds_slice, labels=labels)
pass
```

# Instantiate the Model

```python
model_args = {"backboneID":'ResNet50_pretrained', 
              'RPN_Conv2D_input_filters':512, 
              'RPN_Conv2D_input_kernel_size':(3, 3), 
              'n_anchors':9, 
              'n_classes': len(labels), 
              'RPN_optimizer': tf.keras.optimizers.Adam(learning_rate = 1e-5), 
              'ROI_optimizer': tf.keras.optimizers.Adam(learning_rate = 1e-5), 
              'path_model_dir': None}

model = NN.RCNN.faster.model(**model_args)
```
